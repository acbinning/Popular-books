```python
df.sort_values(by='Approximate sales in millions', ascending=False, inplace=True)
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Approximate sales in millions', y='Book', data=df.head(10))
plt.title('Top 10 books with the highest number of sales')
plt.xlabel('Approximate sales in millions')
plt.ylabel('Book name')
plt.show()
```

```python
df['Original language'].value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Approximate sales in millions', y='Book', hue='Original language', data=df[df['Original language'] != 'English'].head(10))
plt.title('Most popular book in each language and the number of sales')
plt.xlabel('Approximate sales in millions')
plt.ylabel('Book name')
plt.show()
```

```python
# Show the most popular English books and the number of sales, include a key to show the author of each book
plt.figure(figsize=(10,6))
sns.barplot(x='Approximate sales in millions', y='Book', hue='Author(s)', data=df[df['Original language'] == 'English'].head(10))
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Approximate sales in millions', y='Book', data=df[df['Author(s)'] == 'J. K. Rowling'].head(10), palette=['turquoise', 'pink', 'white'], edgecolor='black')
plt.title('Most popular books by J. K. Rowling')
plt.xlabel('Approximate sales in millions')
plt.ylabel('Book name')
plt.show()
```
